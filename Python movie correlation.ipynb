```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
plt.style.use('ggplot')
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
df = pd.read_csv(r'C:\Users\Nugen\Documents\Portfolio\movies.csv')
```

```python
df.head()
```

```python
# Checking for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Checking for duplicates
df.duplicated().sum()
```

```python
# Objective is to see what correlates more to gross earnings
```

```python
# Sort DataFrame by gross earnings
df = df.sort_values('gross', ascending=False)
```

```python
df.head()
```

```python
# Budget vs gross
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'orange'}, line_kws={'color':'blue'})
```

```python
# Scatter plot budget vs gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('budget vs gross earnings')
plt.xlabel('budget')
plt.ylabel('gross')
plt.show()
```

```python
# Correlation Matrix between all numeric columns
df.corr()
```

```python
# Heatmap correlation between all numeric columns
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation matrix for Numeric Features")
plt.show()
```

```python
# Assign values to non-numeric columns
df_convert = df

for col_name in df_convert.columns:
    if(df_convert[col_name].dtype == 'object'):
        df_convert[col_name]= df_convert[col_name].astype('category')
        df_convert[col_name] = df_convert[col_name].cat.codes
        
df_convert
```

```python
# Heatmap correlation between all columns
correlation_matrix = df_convert.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation matrix for Movies")
plt.show()
```

```python
# Correlation Matrix between all columns
df_convert.corr()
```

```python
pd.set_option('display.max_rows', None)
```

```python
#correlation using unstack method
correlation_mat = df_convert.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
#Finding high correlation
high_corr = sorted_pairs[sorted_pairs > 0.5]
high_corr
```

```python
# Result: Votes and budget have the highest correlation to gross earnings
```
